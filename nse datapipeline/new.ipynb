```python
import pandas as pd

# Sample data with date strings
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03'],
       'Open': [100, 105, 110],
       'High': [105, 115, 120],
       'Low': [95, 100, 105],
       'Close': [102, 112, 118]}

df = pd.DataFrame(data)

# Convert the 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

print(df)
```

```python
import yfinance as yf

reliance = yf.download("RELIANCE.NS", start="2023-01-01", end="2023-06-30")
```

```python
reliance['Moving_Average_50'] = reliance['Close'].rolling(window=50).mean()
reliance['Moving_Average_200'] = reliance['Close'].rolling(window=200).mean()
```

```python
import matplotlib.pyplot as plt

reliance['Close'].plot(label='Close')
reliance['Moving_Average_50'].plot(label='Moving Average 50-day')
reliance['Moving_Average_200'].plot(label='Moving Average 200-day')

plt.legend()
plt.show()
```

```python
import yfinance as yf

tatasteel = yf.download("TATASTEEL.NS", start="2023-01-01", end="2023-06-30")
```

```python
tatasteel['Moving_Average_50'] = tatasteel['Close'].rolling(window=50).mean()
tatasteel['Moving_Average_200'] = tatasteel['Close'].rolling(window=200).mean()
```

```python
import matplotlib.pyplot as plt

tatasteel['Close'].plot(label='Close')
tatasteel['Moving_Average_50'].plot(label='Moving Average 50-day')
tatasteel['Moving_Average_200'].plot(label='Moving Average 200-day')

plt.legend()
plt.show()
```

```python
buy_signals = tatasteel['Close'] > tatasteel['Moving_Average_50']
sell_signals = tatasteel['Close'] < tatasteel['Moving_Average_50']

tatasteel['Signal'] = 0
tatasteel.loc[buy_signals, 'Signal'] = 1
tatasteel.loc[sell_signals, 'Signal'] = -1

tatasteel['Strategy_Returns'] = tatasteel['Signal'].shift(1) * tatasteel['Close'].pct_change()
cumulative_returns = (tatasteel['Strategy_Returns'] + 1).cumprod()

plt.plot(cumulative_returns)
plt.show()
```

```python
# Moving Average
tatasteel['Moving_Average_50'] = tatasteel['Close'].rolling(window=50).mean()
tatasteel['Moving_Average_200'] = tatasteel['Close'].rolling(window=200).mean()

# Bollinger Bands
tatasteel['Bollinger_High'] = tatasteel['Close'].rolling(window=20).mean() + 2 * tatasteel['Close'].rolling(window=20).std()
tatasteel['Bollinger_Low'] = tatasteel['Close'].rolling(window=20).mean() - 2 * tatasteel['Close'].rolling(window=20).std()

# EMA
tatasteel['EMA_12'] = tatasteel['Close'].ewm(span=12, adjust=False).mean()
tatasteel['EMA_26'] = tatasteel['Close'].ewm(span=26, adjust=False).mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))

tatasteel['Close'].plot(label='Close')
tatasteel['Moving_Average_50'].plot(label='Moving Average 50-day')
tatasteel['Moving_Average_200'].plot(label='Moving Average 200-day')
tatasteel['Bollinger_High'].plot(label='Bollinger High', linestyle='--')
tatasteel['Bollinger_Low'].plot(label='Bollinger Low', linestyle='--')
tatasteel['EMA_12'].plot(label='EMA 12-day')
tatasteel['EMA_26'].plot(label='EMA 26-day')

plt.legend()
plt.show()
```
